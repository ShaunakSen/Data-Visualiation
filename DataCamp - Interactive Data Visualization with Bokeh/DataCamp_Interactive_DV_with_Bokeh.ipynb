### Bokeh.plotting interface

Glyphs are visula shapes that can be drawn on the screen

- like circles, triangle, sqs, lines, wedges etc.

- they have properties like position, color, transparency etc.




```
import numpy as np
```

```
from bokeh.io import output_file, show, output_notebook # output_notebook displays plots inline in jupyter nbs, output_file can be written to an html file
from bokeh.plotting import figure

plot = figure(plot_width=400, tools='pan,box_zoom')

plot.circle(x=[1,2,3,4,5], y=[8,6,5,2,3])

output_notebook()

show(plot);
```

```
output_file('circle.html')
show(plot)
```

```
plot = figure() # create a new figure
# create circles with same x value, but diff y values and radius
plot.circle(x=10, y=[2,4,8,10], size=[10,20,30,40])

output_notebook()

show(plot)
```

#### Ex 2: A scatter plot with different shapes

By calling multiple glyph functions on the same figure object, we can overlay multiple data sets in the same figure.

In this exercise, you will plot female literacy vs fertility for two different regions, Africa and Latin America. Each set of x and y data has been loaded separately for you as fertility_africa, female_literacy_africa, fertility_latinamerica, and female_literacy_latinamerica.

Your job is to plot the Latin America data with the circle() glyph, and the Africa data with the x() glyph.

figure has already been imported for you from bokeh.plotting.



```
#### create random data
fertility_latinamerica = np.random.randint(low=1, high=6, size=10)
female_literacy_latinamerica = np.random.randint(low=50, high=90, size=10)

fertility_africa = np.random.randint(low=5, high=12, size=8)
female_literacy_africa = np.random.randint(low=20, high=70, size=8)
####


# Create the figure: p
p = figure(x_axis_label='fertility (children per woman)', y_axis_label='female_literacy (% population)')

# Add a circle glyph to the figure p
p.circle(x=fertility_latinamerica, y=female_literacy_latinamerica)

# Add an x glyph to the figure p
p.x(fertility_africa, female_literacy_africa)


# Specify the name of the file
output_file('fert_lit_separate.html')

# Display the plot
show(p)
```
