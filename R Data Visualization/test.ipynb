```r
library(tidyr)
library(ggplot2)
```

```r
temperatures <- data.frame(read.table("temperatures(4).txt",header= T))

head(temperatures)
```

```r
temperatures$Year <- factor(temperatures$Year)
```

```r
meanOfTemperatures <- apply(temperatures[,2:13],1,mean, na.rm=TRUE)
temperatures$mean = meanOfTemperatures
temperatures #part(a)
tail(temperatures)

ranks <- rank(meanOfTemperatures, ties.method= "first")
temperatures$rank <- ranks
head(temperatures)
```

```r
temperatures_long <- gather(temperatures, month_name, temperature_value, Jan:Dec, factor_key = TRUE)

head(temperatures_long)
```

```r
temperatures_long$color_code = ifelse(temperatures_long$rank <= 10, 
                                      "medium",
                                      ifelse(temperatures_long$rank > 130, "high", "low"))

head(temperatures_long)
```

```r
g <- ggplot(data=temperatures_long, aes(x=month_name, y=temperature_value, group=Year)) +
  geom_line(aes(color=color_code))


# g + geom_hline(yintercept=0, color = "black", size=0.5)
```

```r
plot(g)
```

```r
g <- g + scale_color_brewer(palette = "RdBu")

plot(g)
```

```r
limits <- seq(-2, 2, 0.5)

print(limits)

new_limits <- c()
i=1

for (x in c(1:length(limits))){
    if (limits[x] > 0) {
        new_limits[i] <- paste("+", limits[x], "°C", sep = "") 
    }
    else {
        new_limits[i] <- paste(limits[x], "°C", sep = "") 
    }
    
    i <- i+1
}

new_limits[5] <- "20th century average"

new_limits
```

```r
g <- g + scale_y_continuous(breaks = seq(-2, 2, 0.5), labels  = new_limits)

plot(g)
```

```r
g <- g + geom_hline(yintercept=0, color = "black", size=0.5)

g <- g + labs(title="Plot of temperatures for each month of Years: 1880 - 2018", 
              y="Temperature (In °C)", x="Month Name")

plot(g)
```
